```python
import osmnx as ox
import networkx as nx
import folium

# Place choice and route method
place     = "Padua, Italy"
mode      = "walk"           # 'drive', 'bike', 'walk'

# Import geographical information as a graph
graph = ox.graph_from_place(place, network_type=mode)

# Convert to undirected graph (better for plotting)
graph_proj = ox.project_graph(graph)

# Create a folium map centered at Padua
center = [45.4064, 11.8768]  # approximate center of Padua
m = folium.Map(location=center, zoom_start=14, tiles="cartodbpositron")

# Add the graph (nodes + edges) to folium
folium_graph = ox.plot_graph_folium(graph, graph_map=m, popup_attribute="highway", edge_color="blue", edge_width=2)

# Category of places
tags = {'amenity': 'bar'}

# Import (geo)Pandas DataFrame
cafe = ox.geometries_from_place(place, tags=tags)
cafe = cafe.reset_index()
cafe = cafe[cafe["element_type"] == "node"]

# Add cafÃ©s as markers on the map
for idx, row in cafe.iterrows():
    lat, lon = row.geometry.y, row.geometry.x
    folium.CircleMarker(
        location=(lat, lon),
        radius=4,
        color="red",
        fill=True,
        fill_opacity=0.7,
        popup=row.get("name", "Unnamed CafÃ©")
    ).add_to(folium_graph)

# Show map
folium_graph.save("padua_map.html")
print("âœ… Map saved as padua_map.html, open it in your browser.")
```
